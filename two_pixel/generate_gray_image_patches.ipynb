Sample 2 pixel pairs from images used for Fig. 2(a) and (c). 
Images are converted to gray color on the fly.

```python
import numpy as np
import array
import os
import random
from random import randint
import pickle
from skimage import io

import matplotlib.pyplot as plt
```

```python
num_patches_per_file = 2000
# patch_size = 3
patch_height = 1
patch_width = 2
image_folder = '/data/IMAGES/COCO/val2017/'
output_folder = '/data/IMAGES/COCO/'
```

```python
filename = '/data/IMAGES/COCO/val2017/000000209222.jpg'
# filename = '/data/IMAGES/COCO/val2017/000000107339.jpg'
img = io.imread(filename, as_gray=True)  
```

```python
patches = []
for filename in os.listdir(image_folder):
    filename = image_folder + filename
    img = io.imread(filename, as_gray=True)
    if img.max() > 1: # black white image loaded as unit8
        img = img / 255
    H, W = img.shape
    for patch_index in range(num_patches_per_file):
        row = randint(0, H - patch_height)
        col = randint(0, W - patch_width)
        patch = img[row:row + patch_height, col:col + patch_width]
        patches.append(patch)
```

```python
patches = np.stack(patches)
```

```python
patches = patches.astype('float32')
```

```python
patches.shape
```

```python
with open(output_folder + f'coco_10M_gray_{patch_height}x{patch_width}patches.pickle', 'bw') as f:
    pickle.dump(patches, f, -1)
```
